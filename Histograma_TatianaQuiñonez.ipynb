```python
#Librerias 
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

```python
# Ejercicio pregunta 3 examen - Joshue Navarrete 
#Generar la imagen en escala de  grises y el histograma
url2= 'https://placekitten.com/800/571'
myImg = io.imread(url2)
myGrayImg = cv.cvtColor(myImg, cv.COLOR_BGR2GRAY)
cv2_imshow(myGrayImg)
plt.hist(myGrayImg.ravel(),bins = 256, range = [0, 256])
plt.show()
```
