## Generar Histograma

```
url="https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"
image = io.imread(url) 
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
plt.hist(gray_image.ravel(),bins = 256, range = [0, 256])
plt.show()
```
